```python
import numpy as np
import matplotlib.pyplot as plt

# 数据
t = np.arange(1, 7, 1)
lower_bound = [0, 0, 2, 5, 30, 8]
main_line = [1, 2, 3, 6, 45, 9]
upper_bound = [2,4, 6, 8, 50, 10]

# 绘制主线
plt.plot(t, main_line, label='Main Line', color='blue')

# 绘制上下界区间
plt.fill_between(t, lower_bound, upper_bound,
                 color='gray', alpha=0.3, label='Interval')

# 设置图表元素
plt.title('Time Series with Interval')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend(loc='best')

# 显示图表
plt.show()
```
